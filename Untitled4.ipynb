```python
# ===== deps =====
!pip install kagglehub[pandas-datasets] pandas numpy scikit-learn networkx python-dateutil tldextract matplotlib scipy

import re, gc, hashlib
import numpy as np
import pandas as pd
import networkx as nx
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
from dateutil import tz
from email import message_from_string
from email.utils import parsedate_to_datetime
import tldextract
import matplotlib.pyplot as plt
from scipy.stats import beta

# ===== load kaggle =====
import kagglehub
from kagglehub import KaggleDatasetAdapter

FILE_PATH = "emails.csv"  # adjust if needed

df_raw = kagglehub.load_dataset(
    KaggleDatasetAdapter.PANDAS,
    "wcukierski/enron-email-dataset",
    FILE_PATH,
)
print("raw shape:", df_raw.shape)

# ===== normalize =====
raw_col = next((c for c in ["message","raw_message","text","raw"] if c in df_raw.columns), None)
if raw_col is None:
    raise ValueError("No raw message column found")
df_raw = df_raw.rename(columns={raw_col: "raw_message"})

# ===== helpers =====
def parse_addresses(header_val: str):
    # split emails
    if not header_val:
        return []
    parts = re.split(r'[;,]', header_val)
    addrs = []
    for p in parts:
        m = re.search(r'[\w.\-%+]+@[\w.\-]+', p or "")
        if m:
            addrs.append(m.group(0).lower().strip())
    return list(dict.fromkeys(addrs))

def enron_domain(addr: str) -> bool:
    # domain check
    if not addr or "@" not in addr:
        return False
    ext = tldextract.extract(addr)
    dom = ".".join([d for d in [ext.domain, ext.suffix] if d])
    return dom.endswith("enron.com")

def to_utc(dtobj):
    # to utc
    if dtobj is None:
        return None
    if dtobj.tzinfo is None:
        return dtobj.replace(tzinfo=tz.UTC)
    return dtobj.astimezone(tz.UTC)

# ===== parse =====
def fast_parse(row_msg: str):
    # minimal parse
    try:
        em = message_from_string(row_msg or "")
        dt = parsedate_to_datetime(em.get("Date")) if em.get("Date") else None
        dt = to_utc(dt)
        sender = parse_addresses(em.get("From"))[:1]
        to_ = parse_addresses(em.get("To"))
        cc_ = parse_addresses(em.get("Cc"))
        bcc_ = parse_addresses(em.get("Bcc"))
        subj = (em.get("Subject") or "").strip()
        return pd.Series({
            "sent_ts": dt,
            "from_addr": sender[0] if sender else None,
            "to_list": to_,
            "cc_list": cc_,
            "bcc_list": bcc_,
            "subject": subj,
        })
    except Exception:
        return pd.Series({
            "sent_ts": None, "from_addr": None,
            "to_list": [], "cc_list": [], "bcc_list": [], "subject": None
        })

df_parsed = df_raw["raw_message"].apply(fast_parse)
df = pd.concat([df_raw, df_parsed], axis=1)

# ===== ids =====
if "file" in df.columns:
    df["message_id"] = df["file"].astype(str)
else:
    df["message_id"] = df["raw_message"].apply(lambda x: hashlib.md5((x or "").encode()).hexdigest())

# ===== clean =====
df = df.dropna(subset=["sent_ts", "from_addr"]).reset_index(drop=True)
df["sent_ts"] = pd.to_datetime(df["sent_ts"], utc=True)
print("parsed shape:", df.shape)

# ===== edges =====
def explode_edges(df):
    # src->dst
    recips = []
    for _, r in df.iterrows():
        full_list = (r["to_list"] or []) + (r["cc_list"] or [])
        for dst in full_list:
            recips.append((r["message_id"], r["from_addr"], dst, r["sent_ts"]))
    if not recips:
        return pd.DataFrame(columns=["message_id","src","dst","sent_ts"])
    e = pd.DataFrame(recips, columns=["message_id","src","dst","sent_ts"])
    e = e[e["src"] != e["dst"]]
    return e

edges = explode_edges(df).sort_values("sent_ts").reset_index(drop=True)
edges["sent_ts"] = pd.to_datetime(edges["sent_ts"], utc=True)
edges["src_id"] = edges["src"].str.lower()
edges["dst_id"] = edges["dst"].str.lower()

# ===== weeks =====
edges["week_start"] = edges["sent_ts"].dt.to_period("W-MON").dt.start_time
edges["week_start"] = edges["week_start"].dt.tz_localize("UTC")

# ===== rarity =====
pair_freq = edges.groupby(["src_id","dst_id"]).size().rename("pair_n").reset_index()
pair_freq["pair_n_inv"] = 1.0 / pair_freq["pair_n"]
edges = edges.merge(pair_freq, on=["src_id","dst_id"], how="left")

# ===== feats/week =====
def weekly_features(e_week: pd.DataFrame):
    # graph
    G = nx.DiGraph()
    G.add_edges_from(e_week[["src_id","dst_id"]].itertuples(index=False, name=None))
    # deg
    out_deg = dict(G.out_degree())
    in_deg = dict(G.in_degree())
    # clustering
    UG = G.to_undirected()
    clustering = nx.clustering(UG)
    # after-hours
    aw = e_week.copy()
    aw["hour"] = aw["sent_ts"].dt.hour
    aw["is_after"] = ~aw["hour"].between(9, 17)
    # partner stats
    sent_cnt = e_week.groupby("src_id").size().rename("emails_sent")
    recv_cnt = e_week.groupby("dst_id").size().rename("emails_recv")
    partners = e_week.groupby("src_id")["dst_id"].nunique().rename("unique_partners")
    after_ratio = aw.groupby("src_id")["is_after"].mean().rename("after_hours_ratio")
    # external ratio
    e_week["dst_enron"] = e_week["dst_id"].apply(enron_domain)
    ext_ratio = (
        e_week.groupby("src_id")
        .apply(lambda x: (x["dst_enron"] == False).mean())
        .rename("external_ratio")
    )
    # combine
    feats = pd.DataFrame({
        "emails_sent": sent_cnt,
        "emails_recv": recv_cnt,
        "unique_partners": partners,
        "after_hours_ratio": after_ratio,
        "external_ratio": ext_ratio,
    }).fillna(0.0)
    feats["out_deg"] = feats.index.map(lambda k: out_deg.get(k, 0)).astype(float)
    feats["in_deg"]  = feats.index.map(lambda k: in_deg.get(k, 0)).astype(float)
    feats["clustering"] = feats.index.map(lambda k: clustering.get(k, 0.0)).astype(float)
    feats = feats.reset_index().rename(columns={"index":"person_id"})
    return feats

weekly_frames = []
for wk, grp in edges.groupby("week_start", sort=True):
    if grp.shape[0] < 5:
        continue
    f = weekly_features(grp)
    f["week_start"] = wk
    weekly_frames.append(f)

features_week = pd.concat(weekly_frames, ignore_index=True) if weekly_frames else pd.DataFrame(
    columns=["person_id","week_start","emails_sent","emails_recv","unique_partners",
             "after_hours_ratio","external_ratio","out_deg","in_deg","clustering"]
)
if not features_week.empty:
    features_week["week_start"] = pd.to_datetime(features_week["week_start"], utc=True)
print("features_week:", features_week.shape)

# ===== anomaly =====
feat_cols = [
    "emails_sent","emails_recv","unique_partners",
    "after_hours_ratio","external_ratio",
    "out_deg","in_deg","clustering"
]
features_week = features_week.dropna(subset=feat_cols).reset_index(drop=True)

scaler = StandardScaler()
X = scaler.fit_transform(features_week[feat_cols].values)

iso = IsolationForest(n_estimators=300, contamination="auto", random_state=42, n_jobs=-1)
iso.fit(X)
sccores = -iso.decision_function(X)

features_week["anomaly_score"] = sccores
features_week["anomaly_flag"] = iso.predict(X) == -1

# ===== plots =====
plt.figure()
plt.hist(features_week["anomaly_score"].values, bins=50)
plt.title("Anomaly Score Distribution")
plt.xlabel("score")
plt.ylabel("count")
plt.tight_layout()
plt.show()

plt.figure()
weekly_mean = (features_week.groupby("week_start")["anomaly_score"].mean().sort_index())
plt.plot(weekly_mean.index, weekly_mean.values)
plt.title("Weekly Mean Anomaly Score")
plt.xlabel("week")
plt.ylabel("mean score")
plt.tight_layout()
plt.show()

plt.figure()
plt.scatter(features_week["out_deg"].values, features_week["anomaly_score"].values, s=10)
plt.title("Out-Degree vs Anomaly")
plt.xlabel("out_deg")
plt.ylabel("score")
plt.tight_layout()
plt.show()

# ===== event windows =====
EVENT_WINDOWS = [
    dict(name="CA_energy_spike_Jun-Sep_2000", start="2000-06-01", end="2000-09-30"),
    dict(name="CA_volatility_Oct-Dec_2000",   start="2000-10-01", end="2000-12-31"),
    dict(name="CA_blackouts_Jan-Mar_2001",    start="2001-01-01", end="2001-03-31"),
    dict(name="CA_crisis_Apr-Jun_2001",       start="2001-04-01", end="2001-06-30"),
]
events_df = pd.DataFrame(EVENT_WINDOWS)
events_df["start"] = pd.to_datetime(events_df["start"], utc=True)
events_df["end"]   = pd.to_datetime(events_df["end"], utc=True)

fw = features_week.copy()
fw["event_name"] = None
fw["in_event"] = False
for _, ev in events_df.iterrows():
    mask = (fw["week_start"] >= ev["start"]) & (fw["week_start"] <= ev["end"])
    fw.loc[mask, "in_event"] = True
    fw.loc[mask, "event_name"] = ev["name"]
features_week = fw

# ===== pick 5 =====
wk_key = features_week.set_index(["person_id","week_start"])[["anomaly_score","in_event"]]
edges = edges.merge(
    wk_key, left_on=["src_id","week_start"], right_index=True, how="left"
).rename(columns={"anomaly_score":"wk_score"})

edges["hour"] = edges["sent_ts"].dt.hour
edges["is_after"] = ~edges["hour"].between(9, 17)
edges["dst_external"] = ~edges["dst_id"].apply(enron_domain)

edges["rarity"] = edges["pair_n_inv"].fillna(0.0)
if edges["rarity"].max() > 0:
    edges["rarity"] = edges["rarity"] / edges["rarity"].max()

edges["event_weight"] = edges["in_event"].fillna(False).astype(float)

edges["msg_score_part"] = (
    edges["wk_score"].fillna(0.0)
    + 0.6*edges["event_weight"]
    + 0.4*edges["rarity"]
    + 0.25*edges["is_after"].astype(float)
    + 0.25*edges["dst_external"].astype(float)
)

msg_scores = (
    edges.groupby(["message_id","week_start","src_id"])["msg_score_part"]
    .mean()
    .rename("msg_score")
    .reset_index()
)

df_min = df[["message_id","raw_message","subject","from_addr","to_list","cc_list","sent_ts"]].copy()
top_msgs = msg_scores.sort_values("msg_score", ascending=False).head(5)
top_msgs = top_msgs.merge(df_min, on="message_id", how="left")

print("\n=== selected emails (5) ===")
for i, r in top_msgs.iterrows():
    print("\n" + "="*80)
    print(f"[{i}] message_id: {r['message_id']}")
    print(f"week_start: {pd.to_datetime(r['week_start'], utc=True)}")
    print(f"src_id: {r['src_id']}")
    print(f"msg_score: {r['msg_score']:.4f}")
    print("-"*80)
    print(r.get("raw_message",""))
    print("="*80)

# ===== bayes scoring =====
# target proxy = in_event
msg_with_event = msg_scores.merge(
    features_week[["person_id","week_start","in_event"]],
    left_on=["src_id","week_start"],
    right_on=["person_id","week_start"],
    how="left"
).drop(columns=["person_id"])

msg_with_event["in_event"] = msg_with_event["in_event"].fillna(False).astype(int)

# bins
try:
    msg_with_event["bin"] = pd.qcut(msg_with_event["msg_score"], q=10, duplicates="drop")
except ValueError:
    msg_with_event["bin"] = pd.cut(msg_with_event["msg_score"], bins=10, include_lowest=True)

# beta prior
a0, b0 = 1.0, 1.0  # uniform

bayes_table = (
    msg_with_event.groupby("bin")
    .agg(k=("in_event","sum"), n=("in_event","count"), s=("msg_score","mean"))
    .reset_index()
)
bayes_table["post_a"] = a0 + bayes_table["k"]
bayes_table["post_b"] = b0 + (bayes_table["n"] - bayes_table["k"])
bayes_table["p_mean"] = bayes_table["post_a"] / (bayes_table["post_a"] + bayes_table["post_b"])
bayes_table["p_lo"] = beta.ppf(0.05, bayes_table["post_a"], bayes_table["post_b"])
bayes_table["p_hi"] = beta.ppf(0.95, bayes_table["post_a"], bayes_table["post_b"])

# map to messages
msg_with_event = msg_with_event.merge(
    bayes_table[["bin","p_mean","p_lo","p_hi"]],
    on="bin", how="left"
)

# attach to selected
top_msgs = top_msgs.merge(
    msg_with_event[["message_id","p_mean","p_lo","p_hi"]],
    on="message_id", how="left"
)

print("\n=== bayes scores (selected) ===")
for i, r in top_msgs.iterrows():
    print(f"[{i}] msg_id={r['message_id']}  p_mean={r['p_mean']:.3f}  ci90=({r['p_lo']:.3f},{r['p_hi']:.3f})")

# ===== plot bayes =====
plt.figure()
plt.plot(bayes_table["s"], bayes_table["p_mean"], marker="o", linestyle="")
plt.title("Bayes Calibrated Probability")
plt.xlabel("msg_score (mean per bin)")
plt.ylabel("p(in_event)")
plt.tight_layout()
plt.show()

# ===== save =====
features_week.sort_values(["week_start","anomaly_score"], ascending=[True,False]).to_csv(
    "features_person_week_with_scores.csv", index=False
)
top_msgs.to_csv("selected_emails_top5_with_bayes.csv", index=False)
bayes_table.to_csv("bayes_calibration_table.csv", index=False)
print("\nsaved: features_person_week_with_scores.csv")
print("saved: selected_emails_top5_with_bayes.csv")
print("saved: bayes_calibration_table.csv")

# ===== done =====
gc.collect()
```
